```python
import polars as pl

# Menggunakan Polars untuk efisiensi ekstrim
# scan_csv tidak membaca data ke RAM, hanya membuat 'plan'
df = pl.scan_csv("/content/data_talking_data/train.csv") \
    .select([
        pl.col("ip").cast(pl.UInt32),
        pl.col("app").cast(pl.UInt16),
        pl.col("device").cast(pl.UInt16),
        pl.col("os").cast(pl.UInt16),
        pl.col("channel").cast(pl.UInt16),
        pl.col("click_time").str.to_datetime(),
        pl.col("is_attributed").cast(pl.UInt8)
    ]) \
    .head(100_000).collect() # Ambil 1 juta baris pertama untuk eksperimen

print(df.head())
```
